```python
import csv
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
# main datasets (all socio-ecomic and demographic data)
datadir = 'data/'
socecodata = pd.read_csv(datadir+'socio-economic data.csv')
agestructdata = pd.read_csv(datadir+'demographic data.csv')
display(socecodata)
display(agestructdata)
```

```python
# cases of social conflicts to evaluate the similarity (their socio-economic and demographic condition) - so called "local extremums"
socecoextrem = pd.read_excel(datadir+'conflict cases (socio-economic data).xlsx')
ageextrem = pd.read_excel(datadir+'conflict cases (demographic data).xlsx')
```

```python
# dataframe to record final results
rankings = pd.DataFrame()
rankings['oktmo'] = socecodata['oktmo']
rankings['name'] = socecodata['name']
rankings['year'] = socecodata['year']
```

```python
# the ranking method
a = 0
for k in range(len(socecoextrem)):
    
    # calculation of closness in socio-economic space
    sim1 = []
    tmp = 0.0
    for i in range(len(socecodata)):
        tmp = mean_squared_error(socecodata.iloc[i, 5:], socecoextrem.iloc[k, 5:]) #without popsize and saldo
        sim1.append(tmp)

    # calculation of closness in demographic space
    sim2 = []
    female = 0.0
    male = 0.0
    b = 0
    while b < len(agestructdata):
        female = mean_squared_error(agestructdata.iloc[b, 4:18], ageextrem.iloc[a, 4:18])
        male = mean_squared_error(agestructdata.iloc[b + 1, 4:18], ageextrem.iloc[a + 1, 4:18])
        sim2.append(female + male)
        b += 2

    # calculation of similarity metric and sorting the dataset
    sim3 = np.array(sim1) + (np.array(sim2) * 0.5) 
    socecodata['similarity'] = sim3
    socecodata = socecodata.sort_values(by='similarity')

    # scores for ranking based on top-300
    risk = []
    tier1 = [1] * 100
    tier2 = [0.5] * 100
    tier3 = [0.2] * 100
    tier4 = [0] * (len(socecodata) - 300)
    risk = tier1 + tier2 + tier3 + tier4

    # save scores in rankings and go to following conflict case
    socecodata['risk'] = risk
    socecodata = socecodata.sort_values(by=['oktmo', 'year'])
    rankings[str(k + 1)] = socecodata['risk']
    socecodata = socecodata[socecodata.columns.drop('similarity')]
    socecodata = socecodata[socecodata.columns.drop('risk')]
    print('Conflict case №: ' + str(k + 1))
```

```python
# Calculation of overall potential for social conflict (total score)
rankings['oktmo'] = rankings['oktmo'].astype(str)
rankings['year'] = rankings['year'].astype(str)
rankings['total score'] = rankings.sum(axis=1, numeric_only=True)
rankings = rankings.sort_values(by=['total score'], ascending=False)
rankings.to_excel('Potential for social conflict.xlsx', index=False)
display(rankings)
```
